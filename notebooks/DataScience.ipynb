```python
def get_family_tree_network():
    df_edges=pd.read_csv(os.path.join(PATH_DATA,'familytree.csv'))
    df_nodes=pd.read_csv(os.path.join(PATH_DATA,'people.csv'))
    
    G=nx.DiGraph()
    for i,row in df_edges.iterrows():
        G.add_edge(row.person1, row.person2, relationship=row.relationship)
    
    for row in df_nodes.to_dict('records'):
        node=row.pop('person')
        if G.has_node(node):
            for k,v in row.items():
                G.nodes[node][k]=v
            G.nodes[node]['is_family']=True
    return G
    
def get_hierarchical_node_positions(G):
    import pygraphviz as pgv
    G_viz = pgv.AGraph(strict=False, directed=True)
    G_viz.graph_attr['rankdir'] = 'TB'

    for n,d in G.nodes(data=True):
        G_viz.add_node(n,**d)
    for a,b,d in G.edges(data=True):
        G_viz.add_edge(a,b,**d)

    # Use Graphviz to calculate positions
    G_viz.layout(prog='dot')
    node2pos={}
    for node in G_viz.nodes():
        x, y = node.attr['pos'].split(',')
        node2pos[node.name]=(float(x), float(y))
    return node2pos
G=get_family_tree_network()
get_hierarchical_node_positions(G)
```

```python
def get_correspondence_network(incl_family_tree=True, min_num_letters=10):
    G=nx.DiGraph() if not incl_family_tree else get_family_tree_network()
    dfx=df_smpl.query('sender!="" & recipient!=""').groupby(['sender','recipient'])[['num_letters','num_words']].sum().reset_index().query(f'num_letters>={min_num_letters}')
    for i,row in dfx.iterrows():
        if G.has_edge(row.sender, row.recipient):
            G.edges[(row.sender,row.recipient)]['num_letters'] = row.num_letters
            G.edges[(row.sender,row.recipient)]['num_words'] = row.num_words
        elif G.has_node(row.sender) or G.has_node(row.recipient):
            G.add_edge(row.sender, row.recipient, relationship='wrote', num_letters=row.num_letters, num_words=row.num_words)
    return G
```

```python
import pandas as pd
import pygraphviz as pgv

def plot_network():
    G_fam = get_family_tree_network()
    fixed = get_hierarchical_node_positions(G_fam)
    G = get_correspondence_network()
    fixed_nodes = list(fixed.keys())
    pos = nx.spring_layout(G, pos=fixed, fixed=fixed_nodes, k=0.25, iterations=50, weight=None)
    plt.figure(figsize=(15, 10))

    s=pd.Series([d.get('num_letters') for a,b,d in G.edges(data=True) if d.get('num_letters')])
    min_letters,max_letters=s.min(),s.max()
    def scale_edge_width(num_letters, max_width=10, min_width=1):
        if not num_letters: return min_width
        return ((num_letters - min_letters) / (max_letters - min_letters)) * (max_width - min_width) + min_width
    
    # Draw edges with scaled variable thickness
    edge_widths = [scale_edge_width(d.get('num_letters')) for u, v, d in G.edges(data=True)]
    nx.draw_networkx_edges(G, pos, width=edge_widths, alpha=0.7)
    nx.draw_networkx_nodes(G, pos, node_color='lightblue', node_size=3000)
    nx.draw_networkx_labels(G, pos, font_size=8, font_weight='bold')
    
    edge_labels = nx.get_edge_attributes(G, 'label')
    nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels)
    return plt.show()
```

```python
plot_network()
```

```python
figdf = df.groupby(['decade','sender'])[['num_letters','num_words']].sum().reset_index().query('decade>0')
figdf
```

```python
import altair as alt

# Create the Altair chart
chart = alt.Chart(figdf).mark_bar().encode(
    x=alt.X('decade:O', title='Decade'),
    y=alt.Y('num_words:Q', title='Number of Letters'),
    color=alt.Color('sender:N', title='Sender'),
    tooltip=[
        alt.Tooltip('sender:N', title='Sender'),
        alt.Tooltip('num_letters:Q', title='Number of Letters'),
        alt.Tooltip('num_words:Q', title='Number of words'),
        alt.Tooltip('decade:O', title='Decade')
    ]
).properties(
    width=400,
    height=300,
    title='Letters per Decade by Sender'
)
chart
```

```python
df_smpl = df.query('num_words>0 & is_journal==False')
df_smpl
```

```python
df_smpl.sender.value_counts().head(25)
```

```python
def get_stats_by(by='sender'):
    odf=df_smpl.groupby(by)[['num_words','num_letters']].sum().sort_values('num_letters',ascending=False)
    odf['words_per_letter'] = odf.num_words / odf.num_letters
    return odf
```

```python
get_stats_by('sender')
```
